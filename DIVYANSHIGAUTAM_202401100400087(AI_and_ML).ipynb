```python
# ====================================
# 1Ô∏è‚É£ UPLOAD CSV FILE (Google Colab)
# ====================================
from google.colab import files
import pandas as pd
import io

# Upload CSV file
uploaded = files.upload()
filename = next(iter(uploaded))
df = pd.read_csv(io.BytesIO(uploaded[filename]))

# ====================================
# 2Ô∏è‚É£ IMPORT LIBRARIES
# ====================================
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style='whitegrid', palette='Set2')
plt.rcParams['figure.figsize'] = (8, 5)

# ====================================
# 3Ô∏è‚É£ BASIC INFO
# ====================================
print("Dataset Shape:", df.shape)
print("Column Names:", df.columns.tolist())
df.head()

# Identify the target column
target_col = 'target' if 'target' in df.columns else 'output'

# ====================================
# 4Ô∏è‚É£ GRAPH 1: Heart Disease Distribution
# ====================================
sns.countplot(data=df, x=target_col)
plt.title("üíì Heart Disease Distribution")
plt.xlabel("Heart Disease (1 = Yes, 0 = No)")
plt.ylabel("Count")
plt.show()

# ====================================
# 5Ô∏è‚É£ GRAPH 2: Age Distribution
# ====================================
if 'age' in df.columns:
    sns.histplot(df['age'], kde=True, bins=30, color='teal')
    plt.title("üéÇ Age Distribution")
    plt.xlabel("Age")
    plt.ylabel("Frequency")
    plt.show()

# ====================================
# 6Ô∏è‚É£ GRAPH 3: Chest Pain Type vs Heart Disease (if available)
# ====================================
# Chest pain column could be named 'cp'
if 'cp' in df.columns:
    sns.countplot(data=df, x='cp', hue=target_col)
    plt.title("‚ù§Ô∏è Chest Pain Type vs Heart Disease")
    plt.xlabel("Chest Pain Type")
    plt.ylabel("Count")
    plt.legend(title="Heart Disease")
    plt.show()

# ====================================
# 7Ô∏è‚É£ GRAPH 4: Correlation Heatmap
# ====================================
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title("üîó Feature Correlation Heatmap")
plt.show()
```
